# BeautifulSoup 学习
'''
    Beautifulsoup 将复杂HTML 文档转换成一个复杂的树形结构，每个节点都是Python对象，所有对象可以
    归纳为4种

    -Tag  #标签及其内容
    
    -NavigableSoup
    
    -BeautifulSoup
    
    -Comment
'''

```python
from bs4 import BeautifulSoup

import requests

base_url = 'https://movie.douban.com/top250?start=0&filter='


    
```

```python
# 获取数据
# import urllib.request

def get_Html(url):
    headers = {
         'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'
    }

    try:
        response = requests.get(url,headers = headers)

        html = response.content.decode('utf-8')
        
    except urllib.error.URLError as result:
            
            print("出错了")
            
            if hasattr(result,"code"):
                
                print(result.code)   #打印错误 码
            
            if hasattr(result,"reason"):
                
                print(result.reason)  #打印错误 信息
                
    return html
            
            
            
```

```python
data_html = get_Html(base_url)

print(data_html)
```

# BeautifulSoup 解析 数据

```python
bs = BeautifulSoup(data_html,"html.parser")

# print（type(bs)）
```

```python
print(bs.head)
print('='*50)
print(bs.title)
print('='*50)
print(bs.a)
```

# 7、文档的搜索   （常用的）

# (1) find_all()

```python
# 字符串过滤  ：会查找与字符串完全匹配的内容

t_list = bs.find_all("a")   #把所以 a 标签找到

print(t_list)
```

# （2） 正则表达式搜索： 使用search() 方法匹配内容

```python
import re
 
t_list = bs.find_all(re.compile("a"))   #这是包含 a 字母的都找出来，还是按照标签，只要标签包含a,就把标签下的内容找出来

print(t_list)
```

# （3） 方法：传入一个函数（方法），根据函数的要求来搜索(了解)

```python
def name_is_exists(tag):
    
    return tag.has_attr("name")  #包含name 的标签

t_list = bs.find_all(name_is_exists)

# print(t_list)
for item in t_list:
    print(item)
```

# (4) 、kwargs  参数

```python
t_list = bs.find_all(id="doubanapp-tip")

for item in t_list:
    
    print(item)
```

```python
# 注意这里 class_ 参数有一个下划线
t_list = bs.find_all(class_ = True)  

for item in t_list:
    
    print(item)
```

```python
t_list = bs.find_all(href = 'https://movie.douban.com/subject/1292052/')

for item in t_list:
    
    print(item)
```

# (5)、text 参数

```python
t_list = bs.find_all(text='肖申克的救赎')
print(t_list)
```

```python
t_list = bs.find_all(text=['肖申克的救赎','霸王别姬','阿甘正传'])

for item in t_list:
    print(item)
```

```python
t_list = bs.find_all(text = re.compile('\d'))  #\d  表示包含数字的标签
for item in t_list:
    print(item)
```

# 6、limit 参数

```python
# limit 限制查找个数，但是一般都是从前往后找，即限制查找前几个标签
t_list = bs.find_all("a",limit=3)   
for item in t_list:
    print(item)
```

# 7、CSS选择器  select

# （1）通过标签来查找

```python
# 1、通过标签来查找

print(bs.select('title'))
```

# （2）通过类名来查找  点‘.’代表类 class

```python
# 2、通过类名来查找  点‘.’代表类 class

t_list = bs.select('.other')    #class = "other"

print(len(t_list))

for item in t_list:
    
    print(item)
```

# (3) 通过 id 来查找  ‘#’ 代表id

```python
#3、 通过 id 来查找  ‘#’ 代表id

t_list = bs.select('#doubanapp-tip')    # id = "doubanapp-tip"

print(len(t_list))

for item in t_list:
    
    print(item)
```

# (4) 通过属性来查找  a[class=" "]   

```python
# 4、 通过属性来查找
# t_list = bs.select('a[class=""]')    # 通过属性来查找   class = ""  主要a标签的class属性是空

t_list = bs.select('span[class="other"]')    #  <span class="other">

print(len(t_list))

for item in t_list:
    
    print(item)
  
```

# 获取对应标签下的内容的两种方法

```python
#  取对应标签下的内容   item.string

for item in t_list:
    
    print(item.string)
```

```python
#  取对应标签下的内容   item.get_text()
for item in t_list:
    
    print(item.get_text())
```

# （5） 通过子标签来查找  “head>title”

```python
t_list = bs.select("div > span")
for item in t_list:
    print(item)
```

# (6) 通过兄弟标签查找  “.a~.b”  a,b为兄弟标签   注意点‘.’——class

```python
t_list = bs.select(".title ~ .other")    #取title标签的兄弟标签的  内容  

print(len(t_list))

print('第0个内容：',t_list[0].get_text())

print('*'*50)

for item in t_list:
    print(item)
```
